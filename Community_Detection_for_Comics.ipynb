```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
# from networkx.algorithms import community
from community import community_louvain
```

```python
char_path = "/content/drive/MyDrive/Back_UP/hero-network.csv"
df_char = pd.read_csv(char_path)
df_char.head()
```

```python
Source = df_char['Character_1']
Destination = df_char['Character_2']
```

```python
G = nx.Graph()
```

```python
for node1,node2 in zip(Source, Destination):
  if(G.has_node(node1) == False):
    G.add_node(node1)
  
  if(G.has_node(node2) == False):
    G.add_node(node2)
  
  if(node1 != node2):
    G.add_edge(node1,node2)
```

```python
# partition = community.best_partition(G)
partition = community_louvain.best_partition(G)
```

```python
communities = {}
for community_id in set(partition.values()):
    print(f"Community {community_id}:")
    members = [nodes for nodes in partition.keys() if partition[nodes] == community_id]
    index = "Community" + str(community_id)
    communities[index] = members
    print(members)
  
```

```python
for x,y in communities.items():
  print(x,"--> ",y)
```
