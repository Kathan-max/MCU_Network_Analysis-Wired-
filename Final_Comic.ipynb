```python
def jaccard_similarity(s1: str, s2: str)->float:
	"""Computes the Jaccard similarity score between s1 and s2."""
	return len(set(s1) & set(s2)) / len(set(s1) | set(s2))
 
jaccard_similarity('SPIDER-MAN/PETER PAR','SPIDER-MAN/PETER PARKER')*100
```

```python
def similarity(s1:str, s2:str)-> float:
  
  if len(s1)>len(s2):
    l = len(s2)
  else:
    l = len(s1)
  
  if len(s1)>len(s2):
    m = len(s1)
  else:
    m = len(s2)

  

  same = 0
  for i in range(l):
    if(s1[i] == s2[i]):
      same+=1
    else:
      return same
   
  return same

print(similarity('SPIDER-MAN/PETER PAR','SPIDER-MAN/PETER PARKER'))
```

```python
def add_to_Graph(G,node1,node2):
  node1 = str(node1)
  node2 = str(node2)
  if(G.has_node(node1) == False):
    G.add_node(node1)
  
  if(G.has_node(node2) == False):
    G.add_node(node2)

  G.add_edge(node1,node2)

  return G
```

```python
import numpy as np
def sort_dict_values(d):
  keys = list(d.keys())
  values = list(d.values())
  sorted_value_index = np.argsort(values)
  sorted_value_index = list(sorted_value_index)
  sorted_value_index.reverse()
  sorted_dict = {keys[i]: values[i] for i in sorted_value_index}
  return sorted_dict
```
