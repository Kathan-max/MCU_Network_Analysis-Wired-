```python

MAX_LINES_PER_FILE = 150

# input_file_path = 'winter_soldier.txt'
for i in range(len(input_file_path)):
  output_dir_path = '/content/output' + text_mov_arr[i]

  if not os.path.exists(output_dir_path):
      os.makedirs(output_dir_path)

  with open(input_file_path[i], 'r') as input_file:

      lines = input_file.readlines()
      total_lines = len(lines)
      num_output_files = (total_lines + MAX_LINES_PER_FILE - 1) // MAX_LINES_PER_FILE

      for i in range(num_output_files):
          output_file_path = os.path.join(output_dir_path, f'output_{i+1}.txt')
          start_index = i * MAX_LINES_PER_FILE
          end_index = min(start_index + MAX_LINES_PER_FILE, total_lines)
          with open(output_file_path, 'w') as output_file:
              output_file.writelines(lines[start_index:end_index])

!zip -r /content.zip /content

from google.colab import files
files.download('/content.zip')
```
